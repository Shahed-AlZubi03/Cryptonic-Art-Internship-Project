```python
# Columns to drop
columns_to_drop = [
    'belongs_to_collection', 'homepage', 'imdb_id', 'original_title', 
    'overview', 'poster_path', 'production_companies', 'spoken_languages', 
    'status', 'tagline', 'video','release_date'
]

movies_metadata['release_date'] = pd.to_datetime(movies_metadata['release_date'], errors='coerce')
movies_metadata['release_year'] = movies_metadata['release_date'].dt.year

# Drop the columns
movies_df = movies_metadata.drop(columns=columns_to_drop)


# Save cleaned dataset
movies_df.to_csv(r"C:\Users\Shahe\my_server\movies_project\data\movies_data.csv", index=False)

print("Columns dropped successfully!")
print("Remaining columns:", movies_df.columns.tolist())
```

```python
movies_df.isnull().sum()
```

```python
movies_df = movies_df.dropna(subset=['release_year'])
movies_df['release_year'] = movies_df['release_year'].astype(int)
```

```python
# Current column order
cols = ['adult', 'budget', 'genres', 'id', 'original_language', 'popularity',
        'production_countries', 'revenue', 'runtime', 'title', 
        'vote_average', 'vote_count', 'release_year']

# Move 'id' to the first position
cols = ['id'] + [col for col in cols if col != 'id']
movies_df = movies_df[cols]

movies_df.head()
```

```python
import ast
import pandas as pd

# Example: movies_df already loaded
# movies_df['genres'] looks like: "[{'id': 12, 'name': 'Adventure'}, ...]"

def extract_genre_names(genre_str):
    try:
        # Convert string to actual list of dicts
        genre_list = ast.literal_eval(genre_str)
        # Extract the 'name' from each dict
        return [genre['name'] for genre in genre_list]
    except (ValueError, SyntaxError):
        return []  # return empty list if parsing fails

# Apply function to all rows
movies_df['genres'] = movies_df['genres'].apply(extract_genre_names)

# Check result
print(movies_df['genres'].head())
```

```python
import ast
import pandas as pd

# Example: movies_df already loaded
# movies_df['production_countries'] looks like:
# "[{'iso_3166_1': 'GB', 'name': 'United Kingdom'}, ...]"

def extract_country_names(country_str):
    try:
        # Convert string to actual list of dicts
        country_list = ast.literal_eval(country_str)
        # Extract the 'name' from each dict
        return [country['name'] for country in country_list]
    except (ValueError, SyntaxError):
        return []  # return empty list if parsing fails

# Apply function to all rows
movies_df['production_countries'] = movies_df['production_countries'].apply(extract_country_names)

# Check result
print(movies_df['production_countries'].head())
```

## RATING DATA

```python
ratings_data = pd.read_csv(r"C:\Users\Shahe\Downloads\ratings.csv\ratings.csv")
ratings_data.head(10)
```

```python
ratings_df= ratings_data
ratings_df['timestamp'] = pd.to_datetime(ratings_df['timestamp'], unit='s', errors='coerce')

# Save cleaned dataset
ratings_df.to_csv(r"C:\Users\Shahe\my_server\movies_project\data\rating_data.csv", index=False)

print("Columns dropped successfully!")
print("Remaining columns:", ratings_df.columns.tolist())
```

```python
ratings_df.head()
```

```python
ratings_df['movieId'] = ratings_df['movieId'].astype(int)
movies_df['id'] = movies_df['id'].astype(int)

# Now filter
ratings_df = ratings_df[ratings_df['movieId'].isin(movies_df['id'])]

print(f"Ratings filtered! Remaining rows: {len(ratings_df)}")
```

```python
# find which movieIds in ratings are missing from movies:
missing_ids = set(ratings_df['movieId']) - set(movies_df['id'])
print("Movies missing from movies_df:", missing_ids)
```
